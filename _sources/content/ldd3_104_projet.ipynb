### 1.6 Parcours du graphe

La classe `Graphe` est maintenant suffisamment riche pour être exploiter par des techniques d'analyse de graphe.

En théorie des graphes, le parcours de graphe est une technique utilisée pour visiter les sommets de proche en proche à partir d'un sommet initial. Il existe de nombreux algorithmes de parcours de graphe, chacun ayant des caractéristiques et des applications spécifiques. Le [parcours en largeur]() (*breadth first search* en anglais) est par exemple utilisé dans l'algorithme de recherche des plus courts chemins de Dijkstra.
